# 3.preprocessing of data

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
C=pd.DataFrame(ss.fit_transform(B),columns=B.columns)
```

# 4.decide no. of clusters

```python
from sklearn.cluster import KMeans
K=range(2,20,1)
WCSS=[]
for i in K:
    km=KMeans(n_clusters=i)
    model=km.fit(C)
    WCSS.append(model.inertia_)
```

```python
WCSS
```

```python
plt.scatter(K,WCSS,c="red")
plt.plot(K,WCSS,c="blue")
plt.xlabel("no. of clusters-k")
plt.ylabel("WCSS|inertia")
plt.xticks(range(1,20,1))
plt.yticks(range(0,1000,50))
plt.title("elbow curve")
plt.savefig("c:/users/ullah/desktop/DS/project-02/elbow-curve.png")

#no. of clusters decided =8
           
```

# 5.cluster model

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=8)
model=km.fit(C)
model.labels_
```

```python
Q=pd.DataFrame(model.labels_,columns=["cluster"])
Q.groupby(by="cluster")["cluster"].count()
```

# 3A. simple linear Regression

we have data set of loan prediction 
If customer is not eligible for the input required amount and duration
then what can be amount for the given duration By regression model

```python
M=pd.read_csv("c:/users/ullah/Desktop/training_set.csv")
M  #in problem statement clearly says (ammount and durration) so as our X and Y is alreday choosen
```

```python
M1=M[["LoanAmount","Loan_Amount_Term"]]
```

```python
M1.shape
```

```python
M1.isna().sum()
```

```python
replacer(M1)
```

```python
X=M1[["Loan_Amount_Term"]]
Y=M1[["LoanAmount"]] # asked as what else ammount can be given for durration i.e LoanAmount=Y

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=31)
```

```python
print("xtrain",xtrain.shape)
print("xtest",xtest.shape)
print("ytrain",ytrain.shape)
print("ytest",xtest.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
model=lm.fit(xtrain,ytrain)
pred=model.predict(xtest)


from sklearn.metrics import mean_absolute_error,mean_squared_error
print("MAE",round(mean_absolute_error(pred,ytest),2))
print("MSE",round(mean_squared_error(pred,ytest),2))
```

```python
Finaldata = xtest
Finaldata[["Actual Loan Amount"]] = ytest["LoanAmount"]
Finaldata[["Pred Loan Amount"]] = pred
Finaldata[["Err"]] = Finaldata["Pred Loan Amount"] - Finaldata["Actual Loan Amount"]
Finaldata[["AbsErr"]] = abs(Finaldata[["Err"]])
Finaldata[["SquaredErr"]] = Finaldata[["AbsErr"]] * Finaldata[["AbsErr"]]
Finaldata    
```
