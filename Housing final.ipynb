# 12.  Lasso model

```python
w=[]
x=0.100  #100+900=1000 sampleset let consider it
for i in range(0,900,1):
    x=x+0.001
    x=round(x,4)
    w.append(x)
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
tp={"alpha":w}

from sklearn.model_selection import GridSearchCV
cv=GridSearchCV(ls,tp,scoring="neg_mean_absolute_error",cv=4)
cvmodel=cv.fit(xtrain,ytrain)
cvmodel.best_params_
```

```python
from sklearn.linear_model import Lasso
lso=Lasso(alpha=3)
model2=lso.fit(xtrain,ytrain)
 
from sklearn.metrics import mean_absolute_error
#training error
pred=model2.predict(xtrain)
MAE_tr=round(mean_absolute_error(ytrain,pred),5)

#testing error
pred=model2.predict(xtest)
MAE_ts=round(mean_absolute_error(ytest,pred),5)
```
